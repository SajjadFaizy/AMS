```python
#Demonstration: Farbbilder laden verarbeiten und SChwerpunkt berechnen und grafisch darstellen

import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np

imgclr = mpimg.imread('dreieck.png')
print(imgclr.shape)
imgclr = (imgclr*255).astype(int)
plt.imshow(imgclr)
plt.show()

[ny, nx, nz] = imgclr.shape
plt.plot(imgclr[:, int(nx/2)])
plt.show()

imggry = np.round(0.2989*imgclr[:, :, 0], 0) + np.round(0.5870*imgclr[:, :, 1], 0) + np.round(0.1140*imgclr[:, :, 2], 0)
imggry = imggry.astype(int)

plt.set_cmap('gray')
plt.imshow(imggry)
plt.show()



plt.plot(imggry[:, int(nx/2)])
plt.plot(imggry[int(ny/2), :])
plt.show()

binimg = imggry > 50
plt.imshow(binimg)
plt.show()

spunkt = np.zeros(2).astype('int')
spges = int(0)

for k in range(nx):
    for m in range (ny):
        if binimg[m, k]:
            spges += 1
            spunkt += [m, k]
            
spunkt = (spunkt/spges).astype('int')
print([spges, spunkt])

plt.imshow(imgclr)
plt.plot([spunkt[1]], [spunkt[0]], 'o')
plt.show()
```
